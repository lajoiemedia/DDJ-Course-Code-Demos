```python
# Import relevant libraries for use
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_gas_prices = pd.read_excel("data/raw/Gas Prices.xlsx")
df_gas_prices
```

```python
df_gas_prices["CPI Change"] = (df_gas_prices['Consumer Price Index'] - df_gas_prices['Consumer Price Index'][0]) / df_gas_prices['Consumer Price Index'][0]
df_gas_prices
```

```python
df_gas_prices["Adjusted Prices"] = df_gas_prices['Avg Retail Price, Regular Gasoline (dollars per gallon)'] / (1 + df_gas_prices['CPI Change'])
df_gas_prices
```

```python
df_gas_prices.plot(x="DATE", y=["Avg Retail Price, Regular Gasoline (dollars per gallon)","Adjusted Prices"])
```

```python
df_gas_prices.to_csv("data/processed/gasprices.csv", index=False)
```
